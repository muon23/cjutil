```python
from pathlib import Path
import pandas as pd

data_root = Path("../data")
data_file = data_root / "iris" / "iris.csv"
data_df = pd.read_csv(data_file)
data_df
```

```python
import numpy as np
from sklearn.preprocessing import StandardScaler

def scale(df: pd.DataFrame, column: str) -> np.ndarray:
    return StandardScaler().fit_transform(df[column].to_numpy().reshape(-1, 1))

sl = scale(data_df, "sepal_length")
sw = scale(data_df, "sepal_width")
pl = scale(data_df, "petal_length")
pw = scale(data_df, "petal_width")

features = np.hstack((sl, sw, pl, pw))
features.shape
```

```python

labels = data_df["species"].to_numpy().reshape(-1, 1)
labels.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    features, labels, test_size=0.2, stratify=labels
)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.ensemble import GradientBoostingClassifier

model = GradientBoostingClassifier().fit(X_train, y_train.ravel())

score = model.score(X_test, y_test.ravel())
print(score)
```

```python
from sklearn.cluster import DBSCAN

clustered = DBSCAN().fit_predict(features)
clustered
```

```python
from sklearn.metrics import normalized_mutual_info_score

normalized_mutual_info_score(labels.ravel(), clustered)
```
